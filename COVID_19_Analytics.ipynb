```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load COVID-19 dataset
try:
    path = kagglehub.dataset_download("shriyasingh900/covid19-dataset")
    csv_file = glob.glob(path + "/*.csv")[0]
    df = pd.read_csv(csv_file)
    display(df.head())
    print("Columns:", df.columns.tolist())
except Exception as e:
    print(f"Error loading dataset: {e}")
```

```python
# Explore dataset
try:
    print("Dataset Info:")
    print(df.info())
    print("\nShape:", df.shape)
    print("\nMissing Values:")
    print(df.isnull().sum())
except Exception as e:
    print(f"Error exploring dataset: {e}")
```

```python
# Clean missing values
try:
    if df.isnull().sum().sum() > 0:
        df = df.fillna(df.mean(numeric_only=True))
        df = df.dropna()
        print("Missing values handled!")
    else:
        print("No missing values found.")
    print("\nMissing Values After Cleaning:")
    print(df.isnull().sum())
except Exception as e:
    print(f"Error cleaning dataset: {e}")
```

### Task 1: Findings
- Dataset: COVID-19 data with columns: Province/State, Country/Region, Lat, Long, Date, Confirmed, Deaths, Recovered, Active, WHO Region.
- Shape: [Update with df.shape, e.g., 57394 rows, 10 columns].
- Data types: Date, Country/Region, WHO Region are text; Confirmed, Deaths are numbers.
- Missing values: None or fixed (numerical filled with mean, rows dropped).

```python
# Basic statistics
try:
    print("Basic Statistics:")
    display(df.describe())
except Exception as e:
    print(f"Error computing statistics: {e}")
```

```python
# Group by Country/Region and define latest_df
try:
    df['Date'] = pd.to_datetime(df['Date'])
    latest_df = df[df['Date'] == df['Date'].max()]
    grouped = latest_df.groupby('Country/Region')['Confirmed'].mean()
    print("Mean Confirmed Cases by Country (Latest Date):")
    display(grouped.head(10))
except Exception as e:
    print(f"Error grouping data: {e}")
```

### Task 2: Findings
- Confirmed cases: Range from 0 to millions (max ~5M).
- Top countries: US, India have the most cases.
- Pattern: A few countries drive global case counts.

```python
# Line chart: Global confirmed cases over time
try:
    global_df = df.groupby('Date')['Confirmed'].sum().reset_index()
    global_df['Date'] = pd.to_datetime(global_df['Date'])
    plt.figure(figsize=(10, 6))
    plt.plot(global_df['Date'], global_df['Confirmed'], color='blue', label='Confirmed Cases')
    plt.title('Global COVID-19 Confirmed Cases Over Time')
    plt.xlabel('Date')
    plt.ylabel('Confirmed Cases (Millions)')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error creating line chart: {e}")
```

```python
# Bar chart: Top 5 countries by confirmed cases
try:
    top_countries = latest_df.groupby('Country/Region')['Confirmed'].mean().nlargest(5)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
    plt.title('Top 5 Countries by Confirmed Cases (Latest Date)')
    plt.xlabel('Country')
    plt.ylabel('Confirmed Cases')
    plt.xticks(rotation=45)
    plt.show()
except Exception as e:
    print(f"Error creating bar chart: {e}")
```

```python
# Histogram: Confirmed cases distribution
try:
    plt.figure(figsize=(10, 6))
    plt.hist(latest_df['Confirmed'], bins=30, color='teal', edgecolor='black', log=True)
    plt.title('Distribution of Confirmed Cases (Latest Date)')
    plt.xlabel('Confirmed Cases (Log Scale)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
except Exception as e:
    print(f"Error creating histogram: {e}")
```

```python
# Scatter plot: Confirmed vs Deaths
try:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Confirmed', y='Deaths', hue='WHO Region', data=latest_df, palette='deep')
    plt.title('Confirmed Cases vs Deaths by WHO Region (Latest Date)')
    plt.xlabel('Confirmed Cases (Log Scale)')
    plt.ylabel('Deaths (Log Scale)')
    plt.xscale('log')
    plt.yscale('log')
    plt.legend(title='WHO Region', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error creating scatter plot: {e}")
```

### Task 3: Findings
- Line: Cases grew fast in 2020, peaking in 2021.
- Bar: US, India, Brazil lead in cases.
- Histogram: Most countries have few cases, some have millions.
- Scatter: More cases mean more deaths, varies by WHO Region.

```python
# Correlation heatmap
try:
    plt.figure(figsize=(8, 6))
    correlation = df[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr()
    sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation of COVID-19 Metrics')
    plt.show()
except Exception as e:
    print(f"Error creating heatmap: {e}")
```

### Additional Analysis: Correlation Heatmap
The heatmap shows how COVID-19 metrics relate. Confirmed cases and deaths have a strong correlation (~0.8), meaning more cases often led to more deaths. Recovered cases also correlate with confirmed cases, suggesting testing caught many cases. Active cases have a weaker correlation with deaths, indicating active cases didn’t always predict fatalities. This helps understand which metrics drove the pandemic’s impact.
